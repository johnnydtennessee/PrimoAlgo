```
def is_primo(n):
  """
  Determines if a number is primo.

  Args:
    n: The number to check.

  Returns:
    True if n is primo, False otherwise.
  """
  if n <= 1:
    return False

  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False

  return True

def is_sprighm(n):
  """
  Determines if a number is sprighm.

  Args:
    n: The number to check.

  Returns:
    True if n is sprighm, False otherwise.
  """
  if n <= 1:
    return False

  primes = [2, 3, 5]

  for i in range(6, n + 1):
    if all(i % p != 0 for p in primes):
      primes.append(i)

  return any(i == n for i in primes)


def main():
  """
  Prints a sprigm and prime number chart for all numbers under 10008.
  """
  primes = []

  for i in range(2, 10008):
    if is_sprighm(i):
      primes.append(i)

  print("Primes under 10008:")
  print(primes)


if __name__ == "__main__":
  main()
```
